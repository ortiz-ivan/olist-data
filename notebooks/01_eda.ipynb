```python
# --- Configuración e Imports ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configuración visual
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid", palette="pastel")

# Ruta base de los datasets
DATA_PATH = "../data/raw"
```

```python
# Cargar datasets
customers = pd.read_csv(os.path.join(DATA_PATH, "olist_customers_dataset.csv"))
geolocation = pd.read_csv(os.path.join(DATA_PATH, "olist_geolocation_dataset.csv"))
order_items = pd.read_csv(os.path.join(DATA_PATH, "olist_order_items_dataset.csv"))
order_payments = pd.read_csv(os.path.join(DATA_PATH, "olist_order_payments_dataset.csv"))
order_reviews = pd.read_csv(os.path.join(DATA_PATH, "olist_order_reviews_dataset.csv"))
orders = pd.read_csv(os.path.join(DATA_PATH, "olist_orders_dataset.csv"))
products = pd.read_csv(os.path.join(DATA_PATH, "olist_products_dataset.csv"))
sellers = pd.read_csv(os.path.join(DATA_PATH, "olist_sellers_dataset.csv"))
category_translation = pd.read_csv(os.path.join(DATA_PATH, "product_category_name_translation.csv"))
```

```python
# Dimensiones y primeras filas
datasets = {
    "customers": customers,
    "geolocation": geolocation,
    "order_items": order_items,
    "order_payments": order_payments,
    "order_reviews": order_reviews,
    "orders": orders,
    "products": products,
    "sellers": sellers,
    "category_translation": category_translation
}

for name, df in datasets.items():
    print(f"\n{name.upper()} -> Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
    display(df.head(2))
```

```python
for name, df in datasets.items():
    print(f"\n--- {name.upper()} ---")
    print(df.info())
    print("\nValores nulos por columna:")
    print(df.isnull().sum())
```

```python
order_reviews['review_comment_title'] = order_reviews['review_comment_title'].fillna('')
order_reviews['review_comment_message'] = order_reviews['review_comment_message'].fillna('')
```

```python
orders['order_approved_at'] = orders['order_approved_at'].fillna('Unknown')
orders['order_delivered_carrier_date'] = orders['order_delivered_carrier_date'].fillna('Unknown')
orders['order_delivered_customer_date'] = orders['order_delivered_customer_date'].fillna('Unknown')
```

```python
products = products.dropna(subset=['product_category_name', 'product_name_lenght', 
                                   'product_description_lenght', 'product_photos_qty'])

products['product_weight_g'] = products['product_weight_g'].fillna(products['product_weight_g'].mean())
products['product_length_cm'] = products['product_length_cm'].fillna(products['product_length_cm'].mean())
products['product_height_cm'] = products['product_height_cm'].fillna(products['product_height_cm'].mean())
products['product_width_cm'] = products['product_width_cm'].fillna(products['product_width_cm'].mean())
```

```python
for name, df in [('CUSTOMERS', customers),
                 ('GEOLOCATION', geolocation),
                 ('ORDER_ITEMS', order_items),
                 ('ORDER_PAYMENTS', order_payments),
                 ('ORDER_REVIEWS', order_reviews),
                 ('ORDERS', orders),
                 ('PRODUCTS', products),
                 ('SELLERS', sellers),
                 ('CATEGORY_TRANSLATION', category_translation)]:
    print(f'{name} -> valores nulos restantes:\n{df.isnull().sum()}\n')
```

```python
from IPython.display import Markdown as md

md("""
### Relaciones entre datasets Olist

- `orders` es la tabla central (una orden = una compra)
- `order_items` se vincula con `orders` por `order_id`
- `order_items` se vincula con `products` y `sellers`
- `order_payments` también se relaciona por `order_id`
- `order_reviews` contiene las reseñas de las órdenes (`order_id`)
- `customers` se une con `orders` por `customer_id`
- `geolocation` permite vincular ubicaciones (por zip_code_prefix)
- `product_category_name_translation` traduce categorías de productos
""")
```

```python
# Merge de ordenes con clientes
df = orders.merge(customers, on="customer_id", how="left")

# Agregar items, productos, y reseñas
df = df.merge(order_items, on="order_id", how="left")
df = df.merge(products, on="product_id", how="left")
df = df.merge(order_reviews, on="order_id", how="left")

# Traducir categorías de producto
df = df.merge(category_translation, on="product_category_name", how="left")

print("Shape del dataset combinado:", df.shape)
df.head(3)
```

```python
# Distribución de estados de las órdenes
plt.figure(figsize=(8,4))
sns.countplot(x="order_status", data=df, order=df['order_status'].value_counts().index)
plt.title("Distribución de estados de las órdenes")
plt.xticks(rotation=45)
plt.show()

# Número de pedidos por año
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['year'] = df['order_purchase_timestamp'].dt.year

plt.figure(figsize=(6,4))
sns.countplot(x="year", data=df)
plt.title("Pedidos por año")
plt.show()

# Distribución de puntajes de reseñas
plt.figure(figsize=(6,4))
sns.countplot(x="review_score", data=df)
plt.title("Distribución de puntajes de reseñas")
plt.show()
```

```python
# Guardar dataset combinado (procesado inicial)
OUTPUT_PATH = "../data/processed/olist_combined_initial.csv"
df.to_csv(OUTPUT_PATH, index=False)
print(f"Archivo guardado en {OUTPUT_PATH}")
```
